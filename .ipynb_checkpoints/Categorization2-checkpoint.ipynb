```python
# Made a duplicate to analyze negative emotion in general as opposed to just despair
# All code is temporary because it can be optimized a lot

# Outline


# Accesses a file with a large list of keywords relating to negative emotion

# Accesses a file with all responses to the polar bear question

# Constructs a list of keywords and a list of responses

# Finds and reports how many keywords are present in each response

# Displays a list of responses that don't have any keywords for further interpretation

# The purpose of this program is to give a convenient initial filtering of the data
# so that data analysis and checking can be completed quicker. Its practice would also
# emphasize more agreement on how to determine code tallies since everyone would be
# contributing to the keyword list

import pandas as pd

# Building the sadness keywords list
keywordFile = open('NegativeEmotionKeywords.txt')
keywords = []
for line in keywordFile.readlines():
    keywords.append(line[:-1])
keywordFile.close()

# Removing duplicates
keywords = list(set(keywords))

# Building the polar bear responses list
responseFile = open('PolarBearResponses.txt')
responses = []
for line in responseFile:
    responses.append(line[:-1])
responseFile.close()


# List of keyword counts to be used for display later
countList = []

# Populate the repsonsesToKeywords dictionary with the number of keywords 
# each response contians
def countKeywords(keywordList, responses):
    responseNum = 1
    for response in responses:
        currentCount = 0
        for keyword in keywordList:
            if keyword.lower() in response.lower():
                currentCount += 1
        countList.append(currentCount)
        responseNum += 1

countKeywords(keywords, responses)

# Printing the response numbers and the number of keywords found in those responses
def printData(responses, countList):
    index = 0
    while index < len(responses):
        print("Response " + str(index + 1) + ": " + 
              str(countList[index]) + " keywords")
        index += 1

# Creating a pandas dataframe to display the number of keywords in each response
pd.set_option('display.max_rows', None)
data = pd.DataFrame({'Keyword Count': countList}, 
                    index = responses)
display(data)

# Printing out all responses with 0 keywords to double check them
noKeywordResponses = []
index = 0
while index < len(countList):
    if countList[index] == 0:
        noKeywordResponses.append(responses[index])
    index += 1
    
noKeywordData = pd.DataFrame(index = noKeywordResponses)

print("\n\n\nAll responses with no negative keywords:")
display(noKeywordData)
```

```python
# Displaying useful stats

keywordCount = 0
for count in countList:
    if count > 0:
        keywordCount += 1
print("Number of negative responses: " + str(keywordCount))
print("Total responses: " + str(len(responses)))
print("Percentage of responses with a negative emotional response: " + 
      str(100 * round((keywordCount / len(responses)), 2)) + "%")
```
