```python
import pandas as pd

# Building the sadness keywords list
sadKeywordFile = open('SadnessKeywords.txt')
sadnessKeywords = []
for line in sadKeywordFile.readlines():
    sadnessKeywords.append(line[:-1])

# Building the polar bear responses list
polarBearResponseFile = open('PolarBearResponses.txt')
responses = []
for line in polarBearResponseFile:
    responses.append(line[:-1])

# Test response
# responses.append("sad , upset , broken , alone")

# Dictionary of responses mapped to the number of keywords they contain
responsesToKeywords = {}

# List of keyword counts to be used for display later
sadCountList = []

# Initializing the content of the responses_to_keywords dictionary
num = 1
for response in responses:
    responsesToKeywords.update({num:0})
    num += 1

# Populate the repsonsesToKeywords dictionary with the number of keywords 
# each response contians
def countKeywords(keywordList, responses):
    responseNum = 1
    for response in responses:
        currentCount = 0
        for keyword in keywordList:
            if keyword.lower() in response.lower():
                currentCount += 1
        sadCountList.append(currentCount)
        responsesToKeywords[responseNum] = currentCount
        responseNum += 1

countKeywords(sadnessKeywords, responses)
print(responsesToKeywords)

# Printing the response numbers and the number of keywords found in those responses
def printData(responses, countList):
    index = 0
    while index < len(responses):
        # Could show response with responses[index]
        print("Response " + str(index + 1) + ": " + str(countList[index]) + " keywords")
        index += 1
printData(responses, sadCountList)

pd.set_option('display.max_rows', None)
data = pd.DataFrame({'Keyword Count': sadCountList}, 
                    index = responses)
display(data)

sadCount = 0
for count in sadCountList:
    if count > 0:
        sadCount += 1
print("Number of sad responses: " + str(sadCount))
print("Total responses: " + str(len(responses)))
print("Percentage of responses with a sad emotional response: " + 
      str(100 * round((sadCount / len(responses)), 2)) + "%")
                
                
```
